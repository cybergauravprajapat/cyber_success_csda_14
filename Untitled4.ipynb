```python
import requests
from bs4 import BeautifulSoup
import gspread
from oauth2client.service_account import ServiceAccountCredentials

def product_detail_into_googlesheet(title, url, Original_Price,discount_price,discount_percent):
    scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]
    creds = ServiceAccountCredentials.from_json_keyfile_name("flipkartsheetupdate-91acd267c8aa.json", scope)
    client = gspread.authorize(creds)
    sheet = client.open("flipkart").sheet1
    sheet.append_row([title, url, Original_Price,discount_price,discount_percent])
    
def get_flipkart_price(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    title = soup.find('span', class_='VU-ZEz').text 
    original_price = soup.find('div', class_='yRaY8j A6+E6v').text 
    discount_price = soup.find('div', class_='Nx9bqj CxhGGd').text 
    discount_percent = soup.find('div', class_='UkUFwK WW8yVX').text  
    product_detail_into_googlesheet(title,url,int(original_price[1:].replace(",","")),int(discount_price[1:].replace(",","")),int(discount_percent[:2]))

if __name__ == '__main__':
    flipkart_url = input("Enter Flipkart product URL: ")  
    get_flipkart_price(flipkart_url)
```
