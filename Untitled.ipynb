```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd
```

```python
# Define the scope
scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]

# Load credentials from JSON file
creds = ServiceAccountCredentials.from_json_keyfile_name("flipkartsheetupdate-91acd267c8aa.json", scope)

# Authorize client
client = gspread.authorize(creds)

# Open Google Sheet by Name
sheet = client.open("flipkart").sheet1  # "sheet1" refers to the first sheet
```

```python
# Get all data from the sheet
data = sheet.get_all_records()

# Convert to a DataFrame
df = pd.DataFrame(data)

# Display data
print(df)
```

```python
# Update cell (Row 2, Column 3) with "New Value"
_ = sheet.update_cell(3, 1, "New Value")

```
